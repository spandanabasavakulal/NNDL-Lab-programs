```python
#step 1
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, RepeatVector,TimeDistributed,Dense
def sigmoid(x):
  return 1/(1+np.exp(-x))
def tanh(x):
  return np.tanh(x)

#step 2
class LSTMcell:
  def __init__(self, input_size, hidden_size):
    self.input_size = input_size
    self.hidden_size = hidden_size
    # weights for gates
    self.Wf = np.random.randn(hidden_size, hidden_size + input_size) # forget gate
    self.Wi = np.random.randn(hidden_size, hidden_size + input_size) # input gate (corrected typo)
    self.Wc = np.random.randn(hidden_size, hidden_size + input_size) # cell state
    self.Wo = np.random.randn(hidden_size, hidden_size + input_size) # output gate

    # biases
    self.bf = np.zeros((hidden_size, 1))
    self.bi = np.zeros((hidden_size, 1)) # input gate bias
    self.bc = np.zeros((hidden_size, 1)) # cell state bias
    self.bo = np.zeros((hidden_size, 1)) # output gate bias

  def forward(self, x, h_prev, c_prev):
    # concatenate h_prev & input x
    combined = np.vstack((h_prev, x))

    # 1. forget gate
    f_t = sigmoid(np.dot(self.Wf, combined) + self.bf) # Corrected weight name and dot product

    # 2. input gate
    i_t = sigmoid(np.dot(self.Wi, combined) + self.bi) # Corrected weight name and dot product
    c_s = tanh(np.dot(self.Wc, combined) + self.bc) # Corrected weight name and dot product

    # 3. cell state update
    c_t = f_t * c_prev + i_t * c_s

    # 4. output gate
    o_t = sigmoid(np.dot(self.Wo, combined) + self.bo) # Corrected weight name and dot product
    h_t = o_t * tanh(c_t)
    return h_t, c_t, f_t, i_t, o_t

np.random.seed(1)
input_size = 1
hidden_size = 3
sequence_length = 4

# sample input sequence (renamed x to X_sequence for clarity)
X_sequence = [np.random.randn(input_size, 1) for _ in range(sequence_length)]

# initialize LSTM
lstm = LSTMcell(input_size, hidden_size)

# initial states
h = np.zeros((hidden_size, 1))
c = np.zeros((hidden_size, 1)) # Corrected np.zeros call

# forward pass through sequence
print("step by step LSTM EXECUTION:\n")
for t in range(sequence_length):
  current_x = X_sequence[t] # Used X_sequence and renamed to current_x
  h, c, f, i, o = lstm.forward(current_x, h, c)
  print(f"Time step {t+1}") # Corrected f-string formatting
  print("Input:", current_x.flatten())
  print("Forget gate:", f.flatten())
  print("Input gate:", i.flatten())
  print("Output gate:", o.flatten())
  print("hidden state:", h.flatten())
  print("-"*40) # Corrected print string
```
